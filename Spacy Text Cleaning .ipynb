```python
SYMBOL_FILTER = re.compile("[!@#%:;,\.?\'\"]")
CLEAN_UP = re.compile("[^a-z]")
WHITESPCAE_FILTER = re.compile("  *")

def extract_words(x):
    x = " ".join(WHITESPCAE_FILTER.sub(" "," ".join(CLEAN_UP.sub(" "," ".join(SYMBOL_FILTER.sub("",x.lower()).split())).split())).split())
    return x
```
